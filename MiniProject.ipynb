```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
mhsdata = pd.read_csv('Mental Illness Survey.csv')
print("Data type : ", type(mhsdata))
print("Data dims : ", mhsdata.shape)
```

335 Records
40 Variable

```python
print(mhsdata.dtypes)
```

```python
#create backup copy
mhsdata_clean = mhsdata.copy()

#rename column variable
mhsdata_clean.rename(columns={'I identify as having a mental illness':'Mental Illness'}, inplace =True)
mhsdata_clean.rename(columns={'I am unemployed':'Unemployed'}, inplace =True)
mhsdata_clean.rename(columns={'Annual income (including any social welfare programs) in USD':'Income'}, inplace =True)

#making column variable more readable
mhsdata_clean.columns = mhsdata_clean.columns.str.upper()
mhsdata_clean.columns = mhsdata_clean.columns.str.replace(" ", "_")
```

```python
pd.set_option('display.max_columns', None)
```

```python
# remove useless column
cols = list(range(0,9))
mhsdata_clean.drop(mhsdata_clean.columns[cols], axis=1, inplace=True)
```

```python
#replace no and yes to 0 and 1
mhsdata_clean = mhsdata_clean.replace(to_replace=['No', 'Yes'], value=[0, 1])

mhsdata_clean = mhsdata_clean.replace(to_replace=['Some\xa0Masters'], value="Some Masters")

#we assume that NaNs means no
mhsdata_clean = mhsdata_clean.fillna(0)

#remove first row (not counting the variables name)
mhsdata_clean = mhsdata_clean.iloc[1:]
display(mhsdata_clean)
```

```python
income = pd.to_numeric(mhsdata_clean.INCOME)
```

```python
# Define the size of the figure directly in plt.subplots()
fig, ax = plt.subplots(figsize=(20, 6))
sb.countplot(data=mhsdata_clean, x='EDUCATION')
plt.show()
```

```python
education = mhsdata_clean.EDUCATION.to_frame().to_records()
mental = mhsdata_clean.MENTAL_ILLNESS.to_frame().to_records()
income = mhsdata_clean.INCOME.to_frame().to_records()
unemployed = mhsdata_clean.UNEMPLOYED.to_frame().to_records()
gender = mhsdata_clean.GENDER.to_frame().to_records()
age = mhsdata_clean.AGE.to_frame().to_records()
```

```python
display(age)
```

```python
display(income)
```

```python
display(unemployed)
```

```python
aboveSixty, fourtyFiveToSixty, thirtyToFourtyFour, eighteenToTwentyNine = [],[],[],[]

for i in age:
    if i[1] == '> 60':
        aboveSixty.append((i[0], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1]))
    elif i[1] == '45-60':
        fourtyFiveToSixty.append((i[0], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1]))
    elif i[1] == '30-44':
        thirtyToFourtyFour.append((i[0], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1]))
    elif i[1] == '18-29':
        eighteenToTwentyNine.append((i[0], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1]))
```

```python
age_categories = [aboveSixty, fourtyFiveToSixty, thirtyToFourtyFour, eighteenToTwentyNine]
age_title = ['aboveSixty', 'fourtyFiveToSixty', 'thirtyToFourtyFour', 'eighteenToTwentyNine']
j = 0;

for i in age_categories:
    
    x_income_sort = sorted(x[1] for x in i)
    y_unemp = [x[2] for x in i]
    
    sex = [x[3] for x in i]
    
    x_income_sort_int = []
    for i in x_income_sort:
        x_income_sort_int.append(int(i))
    
    umemp_str = []
    for i in range(len(y_unemp)):
        if y_unemp[i] == 1:
            umemp_str.append('Unemployed')
        elif y_unemp[i] == 0:
            umemp_str.append('Employed')
    
    fig, ax = plt.subplots(figsize=(20, 4))
    sb.scatterplot(x = x_income_sort_int, y = umemp_str, 
                    hue=umemp_str, hue_order=("Employed", "Unemployed"), 
                    style=sex)
    
    ax.set(xlabel="Income", ylabel="Employment")
    plt.title(age_title[j])
    
    plt.show()
    
    print(age_title[j], "-- Percent Unemployed: ", sum(y_unemp)/len(y_unemp))
    j = j + 1
    
```

Found the unemployment rate across age group

Now we need to regroup the data based on academic achievement

```python
display(education)
```

```python
masters, bachelor, high_school, some_phd, phd, some_highschool, some_bachelor, some_masters= [], [], [] ,[], [], [], [], []
```

```python
for i in education:
    if i[1] == 'Completed Masters':
        masters.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
    elif i[1] == 'Completed Undergraduate':
        bachelor.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
    elif i[1] == 'High School or GED':
        high_school.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
    elif i[1] == 'Completed Phd':
        phd.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
    elif i[1] == 'Some Phd':
        some_phd.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
    elif i[1] == 'Some highschool':
        some_highschool.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
    elif i[1] == 'Some Undergraduate':
        some_bachelor.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
    else:
        some_masters.append((i[0], mental[i[0]-1][1], income[i[0]-1][1], unemployed[i[0]-1][1], gender[i[0]-1][1], age[i[0]-1][1]))
```

```python
edu_list = [masters, bachelor, high_school, some_phd, phd, some_highschool, some_bachelor, some_masters]

title_list = ["Masters", "Bachelor", "High_school", "Some PHD", "PHD", "Some Highschool", "Some Bachelor", "Some Masters"]
```

```python
j = 0
for i in edu_list:
    x_income_sort = sorted(x[2] for x in i)
    y_mental = [x[1] for x in i]
    unemp = [x[3] for x in i]
    sex = [x[4] for x in i]
    age_grp = [x[5] for x in i]
    
    x_income_sort_int = []
    
    for i in x_income_sort:
        x_income_sort_int.append(int(i))
        
    for i in range(len(unemp)):
        if unemp[i] == 1:
            unemp[1] = "Unemployed"
        elif unemp[i] == 0:
            unemp[i] = "Employed"
            
    fig, ax = plt.subplots(figsize=(20,4))
    sb.scatterplot(x = x_income_sort_int, y = y_mental,
                  hue = unemp, hue_order=('Employed', "Unemployed"),
                  style=sex,
                  size= age_grp, size_order = ('> 60', '45-60', '30-44', '18-29'))
    
    ax.set(xlabel = "Income", ylabel= "Mental Illness")
    plt.title(title_list[j])
    j= j+1
```

```python
mhsdata_clean = mhsdata_clean.iloc[:,:-3]

mhsdata_clean = mhsdata_clean.replace(to_replace=['> 60', '45-60', '30-44', '18-29'], value=[65,52,37,23])

mhsdata_clean = mhsdata_clean.replace(to_replace=['Completed Phd', 'Some Phd', 'Completed Masters', 'Some Masters', 'Completed Undergraduate', 'Some Undergraduate', 'High School or GED', 'Some highschool'], value = [7,6,5,4,3,2,1,0])

mhsdata_clean = mhsdata_clean.replace(['Lack of concentration','Anxiety', 'Depression', 
                   'Obsessive thinking', 'Panic attacks', 'Compulsive behavior', 
                   'Mood swings', 'Tiredness'], 1)

mhsdata_clean = mhsdata_clean.replace(to_replace=['Female', 'Male'], value=[0, 1])

mhsdata_clean.rename(columns={'I_HAVE_ONE_OF_THE_FOLLOWING_ISSUES_IN_ADDITION_TO_MY_ILLNESS':'LACK_OF_CONCENTRATION'}, inplace =True)
mhsdata_clean.rename(columns={'UNNAMED:_28':'ANXIETY'}, inplace =True)
mhsdata_clean.rename(columns={'UNNAMED:_29':'DEPRESSION'}, inplace =True)
mhsdata_clean.rename(columns={'UNNAMED:_30':'OBSESSIVE'}, inplace =True)
mhsdata_clean.rename(columns={'UNNAMED:_31':'MOOD_SWINGS'}, inplace =True)
mhsdata_clean.rename(columns={'UNNAMED:_32':'PANIC_ATTACK'}, inplace =True)
mhsdata_clean.rename(columns={'UNNAMED:_33':'COMPULSIVE'}, inplace =True)
mhsdata_clean.rename(columns={'UNNAMED:_34':'TIREDNESS'}, inplace =True)

display(mhsdata_clean)
```

```python
co_Matrix = mhsdata_clean.corr()
f, ax = plt.subplots(figsize=(12, 9))
sb.heatmap(co_Matrix, vmax=.8, square=True);
plt.show()
```

```python
k = 10 #number of variables for heatmap
cols = co_Matrix.nlargest(k, 'UNEMPLOYED')['UNEMPLOYED'].index

mhsdata_clean[cols] = mhsdata_clean[cols].apply(pd.to_numeric, errors='coerce')

# Calculate correlation coefficients
cm = np.corrcoef(mhsdata_clean[cols].values.T)

sb.set(font_scale=1.25)
hm = sb.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
